```python
#Import Necessary Packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
from sklearn.metrics import mean_squared_error,mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import warnings
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)
warnings.filterwarnings("ignore", category=pd.errors.PerformanceWarning)
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/Movie dataset - Movie dataset.csv')
df=df.drop(['Name','Actor 3','Actor 2'],axis=1) # Drop features with Less Importance
df.head()
```

```python
df.info()
```

--------------------------------*Cleaning Data*--------------------------------

```python
df=df.dropna(subset=['Rating','Genre','Director','Actor 1'],how='any') # Drop rows with missing values
df.info()
```

```python
def clean_year(inp):        # Negative number to Positive number for year
  if inp and inp<0:
    return -1*inp
  return inp


def clean_dur(inp):                # Removes Text from the data for proper numeric value
  if inp and type(inp)==str:
    return int(inp[:-4])
  return inp

def clean_votes(inp):              # Removes Noise from Votes Feature
  if inp:
    return int(inp.replace(',',''))
  return inp
```

```python
df['Year']=df.Year.map(clean_year)
df['Duration']=df.Duration.map(clean_dur)
df['Votes']=df.Votes.map(clean_votes)
```

```python
imputer = KNNImputer(n_neighbors=3,missing_values=np.nan)                 # Impute Values for the Missing cell Using K Nearest Neighbor
aa=imputer.fit_transform(df[['Year','Duration','Rating','Votes']])
df.Duration=aa[:,1:2]
```

```python
df.info()
```

```python
df.head()
```

----------------------------*One Hot Encoding for Categorical Data*----------------------------

```python
# Storing different Genre in a set
con=set()
for x in df.Genre:
  arr=x.split(',')
  for y in arr:
    con.add(y.strip())
con=list(con)
```

```python

for x in con:
  df[x]=0
  for index in df.Genre.index:
    if x in df.Genre[index]:
      df[x][index]=1
    else:
      df[x][index]=0
df.head()
```

```python
con=df.Director.unique()
for x in con:
  df[x]=0
  for index in df.Director.index:
    if x in df.Director[index]:
      df[x][index]=1
      break
df.head()
```

```python
con=df['Actor 1'].unique()
for x in con:
  df[x]=0
  for index in df['Actor 1'].index:
    if x in df['Actor 1'][index]:
      df[x][index]=1
      break
```

```python
df=df.drop(columns=['Actor 1','Genre','Director'],axis=1)
```

```python
df.info()
```

------------------------------------*Split the Data : Train-80%       Test-20%*----------------------------------

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(['Rating'],axis=1),df.Rating,test_size=0.2,random_state=35)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
df.shape
```

```python
# Create the model
model = Sequential([
    Dense(2500, input_dim=5223, activation='relu'),
    Dense(1000, activation='relu'),
    Dense(500, activation='relu'),
    Dense(250, activation='relu'),
    Dense(100, activation='relu'),
    Dense(50, activation='relu'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Print model summary
model.summary()
```

--------------------------------*Train The Model*----------------------------

```python
history=model.fit(x_train,y_train,epochs=100)
```

---------------------------------*Evaluate the model*---------------------------

```python
y_predict=model.predict(x_test)
```

---------------------------*Mean Squared Error*-------------------------

```python
mean_squared_error(y_test,y_predict)
```

---------------------------*Mean Absolute Error*-------------------------

```python
mean_absolute_error(y_test,y_predict)
```
